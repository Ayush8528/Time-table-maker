```python
import pandas as pd

!pip install ortools
from ortools.sat.python import cp_model

def create_schedule():
    model = cp_model.CpModel()

    # --- DATA DEFINITIONS ---
    days = [1, 2, 3]
    slots = ["Morning", "Afternoon"]
    rooms = ["Auditorium", "Room 202", "Conf Hall 1"]

    # Mapping Committees to their assigned Officials
    # (In high-end models, the solver can also assign officials to committees)
    committees = {
        "UNHRC": "Chair B",
        "UNSC": "Vice-Chair C",
        "DISEC": "Chair A",
        "IP": "Chair B",
        "Academic Workshop": "Chair A"
    }

    # Requirement: How many sessions each committee needs
    sessions_needed = {
        "UNHRC": 3,
        "UNSC": 3,
        "DISEC": 3,
        "IP": 2,
        "Academic Workshop": 1
    }

    # --- VARIABLES ---
    # b[committee, day, slot, room] is true if the session happens there
    b = {}
    for c in committees:
        for d in days:
            for s in slots:
                for r in rooms:
                    b[(c, d, s, r)] = model.NewBoolVar(f'{c}_d{d}_s{s}_{r}')

    # --- HARD CONSTRAINTS ---

    # 1. Room Constraint: Each room can host at most one committee per slot
    for d in days:
        for s in slots:
            for r in rooms:
                model.AddAtMostOne(b[(c, d, s, r)] for c in committees)

    # 2. Official Constraint: An official cannot be in two places at once
    # (If multiple committees share an official, they can't happen at the same time)
    for d in days:
        for s in slots:
            for official in set(committees.values()):
                relevant_committees = [c for c, off in committees.items() if off == official]
                model.AddAtMostOne(b[(c, d, s, r)] for c in relevant_committees for r in rooms)

    # 3. Session Requirement: Each committee must meet exactly their required number of times
    for c, count in sessions_needed.items():
        model.Add(sum(b[(c, d, s, r)] for d in days for s in slots for r in rooms) == count)

    # 4. Prompt Constraint: Academic Workshop must be Day 1, Morning
    # We force the sum of all rooms for Workshop at Day 1 AM to be 1
    model.Add(sum(b[("Academic Workshop", 1, "Morning", r)] for r in rooms) == 1)

    # 5. Prompt Constraint: Room 202 unavailable after 3:00 PM (Afternoon) on Day 2
    for c in committees:
        model.Add(b[(c, 2, "Afternoon", "Room 202")] == 0)

    # 6. Prompt Constraint: Chair A cannot attend Day 3 Afternoon
    chair_a_committees = [c for c, off in committees.items() if off == "Chair A"]
    for c in chair_a_committees:
        for r in rooms:
            model.Add(b[(c, 3, "Afternoon", r)] == 0)

    # --- SOFT CONSTRAINTS / OPTIMIZATION ---
    # Objective: Try to distribute sessions across rooms evenly (Optional)
    # For this example, we just look for a feasible solution

    # --- SOLVER ---
    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        output = []
        for d in days:
            for s in slots:
                for r in rooms:
                    for c in committees:
                        if solver.Value(b[(c, d, s, r)]):
                            output.append({
                                "Day": f"Day {d}",
                                "Slot": s,
                                "Room": r,
                                "Committee": c,
                                "Official": committees[c]
                            })

        # Display as a pretty Table
        df = pd.DataFrame(output)
        pivot_df = df.pivot(index=['Day', 'Slot'], columns='Room', values='Committee').fillna("-")
        return pivot_df
    else:
        return "No solution found that satisfies all constraints."

# Run and Print
result = create_schedule()
print(result)
```
